```python
import pandas as pd
import plotly.graph_objects as go
df = pd.read_csv("prueba.csv")
```

```python
from decimal import Decimal, getcontext
import math

def calculate_triangle_vertices(start_point, angle_degrees, length, aperture_degrees):
    getcontext().prec = 28  # Establece la precisión de los decimales
    
    angle_radians = Decimal(math.radians(angle_degrees))
    aperture_radians = Decimal(math.radians(aperture_degrees))
    
    second_vertex_lat = Decimal(start_point[0]) + Decimal(length) * Decimal(math.sin(angle_radians))
    second_vertex_lon = Decimal(start_point[1]) - Decimal(length) * Decimal(math.cos(angle_radians))
    dist_start_to_second = Decimal(((second_vertex_lat - Decimal(start_point[0]))**2 + (second_vertex_lon - Decimal(start_point[1]))**2).sqrt())
    
    half_aperture = aperture_radians
    bisector_angle = angle_radians - half_aperture
    
    third_vertex_lat = Decimal(start_point[0]) + Decimal(length) * Decimal(math.sin(bisector_angle))
    third_vertex_lon = Decimal(start_point[1]) - Decimal(length) * Decimal(math.cos(bisector_angle))
    dist_start_to_third = Decimal(((third_vertex_lat - Decimal(start_point[0]))**2 + (third_vertex_lon - Decimal(start_point[1]))**2).sqrt())
    print("Distancia entre start_point y second_vertex:", dist_start_to_second)
    print("Distancia entre start_point y third_vertex:", dist_start_to_third)

    return [
        [float(second_vertex_lat), float(second_vertex_lon)],
        start_point,
        [float(third_vertex_lat), float(third_vertex_lon)]
    ]

start_point = [20.648888888888887, -87.06388888888888]
angle_degrees = 0
length = 0.0090 
aperture_degrees = 45

calculate_triangle_vertices(start_point, angle_degrees, length, aperture_degrees)
```

```python

# Encontrar los nombres o destinos con la mayor cantidad
top_destinos = df.groupby('Nombre')['Cantidad'].sum().nlargest(10)  # Cambiar 'Nombre' a 'Destino' si se prefiere utilizar el destino

# Crear el gráfico de barras
fig = go.Figure([go.Bar(x=top_destinos.index, y=top_destinos.values, marker_color='rgb(158,202,225)')])

# Configurar el diseño del gráfico
fig.update_layout(title={'text': 'Top 10 Destinos por Cantidad', 'x': 0.5, 'y': 0.95, 'xanchor': 'center', 'yanchor': 'top'},
                  xaxis_title='Destino',
                  yaxis_title='Cantidad',
                  plot_bgcolor='rgba(0,0,0,0)',
                  paper_bgcolor='rgba(0,0,0,0)',
                  font=dict(color='black'),
                  bargap=0.15,
                  bargroupgap=0.1)

# Mostrar el gráfico
fig.show()
```
